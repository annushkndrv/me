```python
import numpy as np
import pandas as pd

from tqdm.notebook import tqdm
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import mean_squared_error as mse
from sklearn.tree import plot_tree

import scipy

sns.set(font_scale=1.8, palette='Set2')
```

# Простая часть

Сгенерируем данные.

```python
np.random.seed(42)
X_train = scipy.stats.uniform.rvs(loc=-3, scale = 6, size=(500, 1))
y_train = np.ravel(np.sin(X_train)) + scipy.stats.norm.rvs(scale=0.1, size=500)

plt.figure(figsize=(10, 8))
plt.scatter(X_train, y_train)

plt.show()
```

```python
X_train.shape, y_train.shape
```

Обучите градиентный бустинг `GradientBoostingRegressor` с параметром `learning_rate=1` и 10 деревьями.

```python
regressor = GradientBoostingRegressor(learning_rate=1, n_estimators=10)
regressor.fit(X_train, y_train)
```

Выведите параметры бустинга. Какие деревья используются в бустинге по умолчанию?

```python
print("Параметры бустинга:")
print(regressor.get_params())
```

**Ответ** обычно используются деревья с max_depth = 3 (глубиной), min_samples_leaf = 1, min_samples_split = 2.

Нарисуйте первые 5 деревьев бустинга. Какие они?

```python
regressor.estimators_.shape
```

```python
def plot_tree_with_graphviz(tree, tree_index):
    plt.figure(figsize=(12, 6))
    plot_tree(tree, filled=True, rounded=True, fontsize=10)
    plt.title(f"Tree {tree_index}")
    plt.show()
```

```python
for i, tree in enumerate(regressor.estimators_[:5, 0]):
    plot_tree_with_graphviz(tree, i + 1)
```

**Ответ** деревья, как выяснилось ранее, имеют глубину 3, и разделяются на 2 листа.

У бустинга есть метод `staged_predict`. Он выдает предсказания композиции первых $i$ моделей бустинга. Используя этот метод, нарисуйте для каждого  $i$ два графика:
* График предсказаний первых $i$ моделей и истинных ответов;
* График ошибки на шаге $i$ и предсказаний $i$-й компоненты бустинга.

```python
staged = list(regressor.staged_predict(X_train))

for i in range(10):
    if i == 0:
      previous_predict = 0
    else:
      previous_predict = staged[i-1]

    error = y_train - staged[i]
    current_step_predict = staged[i] - previous_predict

    # построение графиков
    plt.figure(figsize=(10, 6))
    plt.scatter(X_train, y_train, label='Истинные ответы', alpha=0.5)
    plt.scatter(X_train, staged[i], label='Предсказания', alpha=0.5)
    plt.title(f'Предсказания первых {i+1} моделей и истинные ответы')
    plt.legend()
    plt.show()


    plt.figure(figsize=(10, 6))
    plt.scatter(X_train, error, label='Ошибка', alpha=0.5)
    plt.scatter(X_train, current_step_predict, label='Предсказания i-й компоненты', alpha=0.5)
    plt.title(f'Ошибка на шаге {i+1} и предсказания {i+1}-й компоненты бустинга')
    plt.legend()
    plt.show()
```

Сделайте вывод. Как выглядят графики на разных шагах? Как обучается бустинг?
