```python
def plot_distributions(data, title, ax):
    sns.histplot(data, kde=False, bins=30, color='blue', alpha=0.5, ax=ax, stat='density')
    sns.kdeplot(data, color='red', ax=ax)
    if title in ['pBRAF_N', 'pCREB_N']:
        x = np.linspace(data.min(), data.max(), 1000)
        if title == 'pBRAF_N':
            y = sps.norm.pdf(x, mu_pBRAF_N, sigma_pBRAF_N)
        else:
            y = sps.norm.pdf(x, mu_pCREB_N, sigma_pCREB_N)
        ax.plot(x, y, 'g--', label='Norm PDF')
    else:
        x = np.linspace(data.min(), data.max(), 1000)
        if title == 'ITSN1_N':
            y = sps.gamma.pdf(x, shape_ITSN1_N, loc=loc_ITSN1_N, scale=scale_ITSN1_N)
        else:
            y = sps.gamma.pdf(x, shape_DYRK1A_N, loc=loc_DYRK1A_N, scale=scale_DYRK1A_N)
        ax.plot(x, y, 'g--', label='Gamma PDF')
    ax.set_title(title)
    ax.legend()


fig, axes = plt.subplots(2, 2, figsize=(15, 10))

plot_distributions(ITSN1_N_new, 'ITSN1_N', axes[0, 0])
plot_distributions(DYRK1A_N_new, 'DYRK1A_N', axes[0, 1])
plot_distributions(pBRAF_N_new, 'pBRAF_N', axes[1, 0])
plot_distributions(pCREB_N_new, 'pCREB_N', axes[1, 1])

plt.subplots_adjust(wspace=0.3, hspace=0.3)
```

Реализуйте функции, возвращающие значения функции правдоподобия  и логарифмической функции правдоподобия для гамма распределения с параметрам $\theta, \beta$. Функция принимает реализацию выборки и параметры. Логарифмическую функцию реализуйте с использованием формулы после ее упрощения, а не как взятие функции логарифма от уже вычисленного значения функции правдоподобия.

Методами `pdf` и `logpdf` пользоваться запрещено. Для подсчета гамма-функции используйте `scipy.special.gamma`, а для подсчета логарифма гамма-функции &mdash; `scipy.special.loggamma`.


Выведите следующие значения:
- функции правдоподобия;
- логарифма, взятого от функции правдоподобия;
- логарифмической функции правдоподобия

на следующих реализациях выборок:
- `np.ones(5) * 5` &mdash; 5 пятерок;
- `np.ones(500) * 5` &mdash; 500 пятерок.

Параметр $\beta$ положите равным 10, параметр $\theta$ положите равным 2.

Сделайте выводы. Какую функцию лучше вычислять на практике и почему?


```python
def gamma_likelihood(x, theta, beta):
    n = len(x)
    likelihood = np.prod((theta ** beta) * (x ** (beta - 1)) * np.exp(-theta * x) / sps.gamma(beta))
    return likelihood

def gamma_log_likelihood(x, theta, beta):
    n = len(x)
    log_likelihood = np.sum(beta * np.log(theta) + (beta - 1) * np.log(x) - theta * x - sps.loggamma(beta))
    return log_likelihood


theta = 2
beta = 10


sample1 = np.ones(5) * 5
sample2 = np.ones(500) * 5


likelihood1 = gamma_likelihood(sample1, theta, beta)
log_likelihood1 = gamma_log_likelihood(sample1, theta, beta)
log_likelihood_from_likelihood1 = np.log(likelihood1)

likelihood2 = gamma_likelihood(sample2, theta, beta)
log_likelihood2 = gamma_log_likelihood(sample2, theta, beta)
log_likelihood_from_likelihood2 = np.log(likelihood2)


print(f"Функция правдоподобия для sample1: {likelihood1}")
print(f"Логарифм функции правдоподобия для sample1: {log_likelihood_from_likelihood1}")
print(f"Логарифмическая функция правдоподобия для sample1: {log_likelihood1}")

print(f"Функция правдоподобия для sample2: {likelihood2}")
print(f"Логарифм функции правдоподобия для sample2: {log_likelihood_from_likelihood2}")
print(f"Логарифмическая функция правдоподобия для sample2: {log_likelihood2}")
```

**Вывод**
для маленькой выборки значение функции правдоподобия  нормальная.
для большой выборки (500 наблюдений) значение функции правдоподобия очень мало. это связано с тем, что произведение многих маленьких чисел (плотностей вероятностей) быстро стремится к нулю.

логарифмическая функция правдоподобия для маленькой выборки и для большой выборки практически совпадают с логарифмом функции правдоподобия.

## Задача 5.

Смоделируем шум разными способами и посмотрим, как проявляется робастность оценок и как она связана с их асимптотической дисперсией. Будем сравнивать между собой:
- выборочное среднее, $\overline{X}$
- выборочную медиану, $\hat{\mu}$
- усеченное среднее, $\overline{X}_{\alpha}$ (фукнция `sps.trim_mean()`)

В качестве базовой модели будем рассматривать $N(\theta, 1)$, для которого нам нужно оценить параметр $\theta$. Однако, как бывает часто в жизни, нам приходят не чистые измерения, а с некоторым шумом, который может иметь разную природу. Чтобы его смоделировать, сгенерируем выборку из смешанного распределения следующим образом:
1. Сгенерируем выборку из нашего распределения $N(\theta, 1)$;
2. Сгенерируем выборки из распределения, которое задает шум (см. примеры ниже);
3. Сгенерируем монетку из распределения Бернулли, которая дает 1 с вероятностью $\varepsilon$, где $\varepsilon \in (0, 0.5)$;
4. На тех позициях, где выпал 0, возьмем данные из $N(\theta, 1)$, где 1 - из шумовых данных.

Например,

```python
theta = 0
sample_count = 10000
sample_size = 100

# Создаем истинное распределение данных и шума
X_base = sps.norm(loc=theta).rvs(size=(sample_count, sample_size))
X_outlier = sps.norm(scale=3).rvs(size=(sample_count, sample_size))

# Создаем монетку, которая будет определять долю шума в данных
eps = 0.2
coin = sps.bernoulli(p=eps).rvs(size=sample_size)

X = np.where(coin, X_outlier, X_base)
X.shape
```

```python
grid = np.linspace(-4, 4, 100)

plt.figure(figsize=(12, 4))

plt.hist(X[0, :], bins=50, density=True)
plt.plot(grid, sps.norm.pdf(grid), label="$N(0, 1)$")

plt.title("Распределение данных с шумом")
plt.legend();
```

Вам нужно будет сравнить между собой оценки по их смещенности и дисперсии для фиксированного размера выборки:
1. Сгенерируйте зашумленную выборку размера (10000, 100);

2. Для каждой из оценок посчитайте смещение (математического ожидание оценки минус истинной значение параметра), а также ее дисперсию;

3. Для сравнения оценок между собой посчитайте отношение дисперсии для выборочного срднего и дисперсии других оценок, то есть найдите две величины: $\hat{\sigma}^2_{\overline{X}} / \hat{\sigma}^2_{\overline{X}_{\alpha}}$ и $\hat{\sigma}^2_{\overline{X}} / \hat{\sigma}^2_{\hat{\mu}}$;

4. Повторите первые шаги для не менее сотни значений $\varepsilon \in (0, 0.5)$;

5. Нарисуйте зависимость смещения и отношения дисперсий от доли шума, каждую на своем графике.

Эту процедуру нужно провести для трех видов шума и сравнить оценки с точки зрения их разброса и точности предсказаний:
1. $N(10, 1)$
2. $N(0, 9)$
3. $Cauchy(0)$

*Результаты могут получиться шумными, поэтому вместе с ними также отрисовывайте на графике сглаженные кривые, посчитанные с помощью скользящего среднего (может пригодиться функция `np.concolve()`)*

```python
theta = 0
sample_count = 10000
sample_size = 100
eps_values = np.linspace(0.01, 0.5, 100)
```

```python
def estimate_parameters(X):
    mean_est = np.mean(X, axis=1)
    median_est = np.median(X, axis=1)
    trim_mean_est = sps.trim_mean(X, 0.1, axis=1)
    return mean_est, median_est, trim_mean_est

def calculate_bias_variance(estimates, true_value):
    bias = np.mean(estimates) - true_value
    variance = np.var(estimates)
    return bias, variance

def generate_noisy_sample(theta, sample_count, sample_size, eps, noise_dist):
    X_base = sps.norm(loc=theta).rvs(size=(sample_count, sample_size))
    X_outlier = noise_dist.rvs(size=(sample_count, sample_size))
    coin = sps.bernoulli(p=eps).rvs(size=(sample_count, sample_size))
    X = np.where(coin, X_outlier, X_base)
    return X
```

```python
bias_mean = []
bias_median = []
bias_trim_mean = []
variance_mean = []
variance_median = []
variance_trim_mean = []


for eps in eps_values:
    noise_dists = [sps.norm(loc=10, scale=1), sps.norm(loc=0, scale=9), sps.cauchy(loc=0)]
    for noise_dist in noise_dists:
        X = generate_noisy_sample(theta, sample_count, sample_size, eps, noise_dist)
        mean_est, median_est, trim_mean_est = estimate_parameters(X)

        bias_mean.append(calculate_bias_variance(mean_est, theta)[0])
        bias_median.append(calculate_bias_variance(median_est, theta)[0])
        bias_trim_mean.append(calculate_bias_variance(trim_mean_est, theta)[0])

        variance_mean.append(calculate_bias_variance(mean_est, theta)[1])
        variance_median.append(calculate_bias_variance(median_est, theta)[1])
        variance_trim_mean.append(calculate_bias_variance(trim_mean_est, theta)[1])
```

```python
bias_mean = np.array(bias_mean).reshape(3, -1)
bias_median = np.array(bias_median).reshape(3, -1)
bias_trim_mean = np.array(bias_trim_mean).reshape(3, -1)
variance_mean = np.array(variance_mean).reshape(3, -1)
variance_median = np.array(variance_median).reshape(3, -1)
variance_trim_mean = np.array(variance_trim_mean).reshape(3, -1)

def smooth(y, box_pts):
    box = np.ones(box_pts) / box_pts
    y_smooth = np.convolve(y, box, mode='same')
    return y_smooth
```
