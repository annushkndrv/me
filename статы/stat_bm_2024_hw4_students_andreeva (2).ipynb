## Задача 5.

Пусть $X_1, \ldots, X_n$ &mdash; выборка из распределения $\mathcal{N}(a, \sigma^2)$. Визуализируйте свойства состоятельности и асимптотической нормальности для оценки $S^2$ (см. задачу 1) дисперсии $\sigma^2$.


```python
a = 0
sigma2 = 1
sample_count = 500
sample_size = 100
samples = np.random.normal(a, np.sqrt(sigma2), size=(sample_count, sample_size))
```

```python
estimation = np.array([np.var(samples[:, :i], axis=1) for i in range(1, sample_size+1)]).T
```

```python
def est_plot(estimation, sample_count, sample_size, left=None, right=None, xlim = sample_size):
    '''
    estimation -- массив оценок от размера выборки
    sample_count -- количество генерируемых выборок
    sample_size -- размер каждой выборки
    left, rigth -- границы доверительного интеравала, нужны будут далее
    xlim = область по x
    '''

    plt.figure(figsize=(15, 7))
    for i in range(sample_count):
        plt.plot(np.arange(sample_size) + 1, estimation[i], color='blue', alpha=0.05)
    if type(left) and type(right) is np.ndarray:

        plt.plot(np.arange(sample_size) + 1, left, color='black')
        plt.plot(np.arange(sample_size) + 1, right, color='black')
        labels = [r'$S^2$', r'$\sigma$',
                  'Границы доверительного интервала']
        handels = [Line2D([0], [0], color='blue', lw=2),
                   Line2D([0], [0], color='red', lw=2),
                   Line2D([0], [0], color='black', lw=2)]

    else:
        labels = [r'$S^2$', r'$\sigma$']
        handels = [Line2D([0], [0], color='blue', lw=2),
                   Line2D([0], [0], color='red', lw=2),]

    plt.hlines(np.var(samples), 0 ,sample_size, color='red')
    plt.title('Зависимость оценки от размера выборки')
    plt.xlabel('Размер выборки')
    plt.ylim((0, 5))
    plt.legend(handels, labels)
    plt.xlim((0, xlim));
```

```python
from matplotlib.lines import Line2D
sigma2 = 1
est_plot(estimation, sample_count, sample_size)
```

```python
right = np.var(samples) + 3*np.var(samples)/np.sqrt(np.arange(sample_size)+1)
left = np.var(samples) - 3*np.var(samples)/np.sqrt(np.arange(sample_size)+1)
est_plot(estimation, sample_count, sample_size, right=right, left=left)
```

Судя по графикам оценка является несмещенной (стремится к среднего при увеличении выборки) и состоятельность (вероятность выхода за пределы доверительного интервала с увеличением выборки уменьшается).
