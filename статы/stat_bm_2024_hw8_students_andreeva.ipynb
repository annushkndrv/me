Построим для этих данных доверительные интервалы. Так как мы не знаем распределения, из которого нам приходят результаты эксперимента, воспользуемся бутстрепом.

**1.** Реализуйте подсчет массива бутстрепных статистик для данной выборки и вычисление трех доверительных интервалов.

```python
def bootstrap(sample, estimator, B_sample_size=None, B=10_000):
    '''
    Строит бустрепные оценки по реализации выборки

    :param sample: выборка
    :param estimator: функция подсчета оценки
    :param B_sample_size: размер бустрепной выборки,
                          если None, то равен размеру исходной выборки
    :param B: число бустрепных выборок
    '''
    if B_sample_size is None:
        B_sample_size = len(sample)

    estimates = np.zeros(B)

    for i in range(B):
        bootstrap_sample = np.random.choice(sample, size=B_sample_size, replace=True)
        estimates[i] = estimator(bootstrap_sample)

    return estimates
```

```python
def bootstrap_normal_confidence_interval(theta, bootstrap_estimations, alpha=0.95):
    '''
    Строит левую и правую границы нормального бутстрепного доверительного интервала

    :param theta: значение статистики на выборке
    :param bootstrap_estimations: бутстрепные статистики
    :param alpha: уровень доверия
    '''
    std_estimate = np.std(bootstrap_estimations)
    z_critical = sps.norm.ppf((1 + alpha) / 2)

    lower_bound = theta - z_critical * std_estimate
    upper_bound = theta + z_critical * std_estimate

    return lower_bound, upper_bound


def bootstrap_central_confidence_interval(theta, bootstrap_estimations, alpha=0.95):
    '''
    Строит левую и правую границы центрального бутстрепного доверительного интервала

    :param theta: значение статистики на выборке
    :param bootstrap_estimations: бутстрепные статистики
    :param alpha: уровень доверия
    '''
    sorted_estimates = np.sort(bootstrap_estimations)
    lower_index = int(np.ceil((alpha / 2) * len(sorted_estimates)))
    upper_index = int((1 - alpha / 2) * len(sorted_estimates))

    lower_bound = sorted_estimates[lower_index]
    upper_bound = sorted_estimates[upper_index]

    return lower_bound, upper_bound


def bootstrap_quantile_confidence_interval(theta, bootstrap_estimations, alpha=0.95):
    '''
    Строит левую и правую границы квантильного бутстрепного доверительного интервала

    :param theta: значение статистики на выборке
    :param bootstrap_estimations: бутстрепные статистики
    :param alpha: уровень доверия
    '''
    lower_bound = (np.percentile(bootstrap_estimations, (alpha / 2) * 100))
    upper_bound = (np.percentile(bootstrap_estimations, (1 - alpha / 2) * 100))

    return lower_bound, upper_bound
```

**2.** Для каждого признака постройте бутстрепные доверительные интервалы для $\theta = EX_1$. Результаты запишите в виде таблицы. Сделайте выводы.

```python
df_ci = pd.DataFrame(
                     columns=data.columns)
def estimator(sample):
    return sample[0]

for protein in proteins:
    sample = data[protein].values
    theta = estimator(sample)
    bootstrap_estimations = bootstrap(sample, estimator)

    normal_lower, normal_upper = bootstrap_normal_confidence_interval(theta, bootstrap_estimations)
    central_lower, central_upper = bootstrap_central_confidence_interval(theta, bootstrap_estimations)
    quantile_lower, quantile_upper = bootstrap_quantile_confidence_interval(theta, bootstrap_estimations)

    df_ci.loc['normal_lower', protein] = normal_lower
    df_ci.loc['normal_upper', protein] = normal_upper
    df_ci.loc['central_lower', protein] = central_lower
    df_ci.loc['central_upper', protein] = central_upper
    df_ci.loc['quantile_lower', protein] = quantile_lower
    df_ci.loc['quantile_upper', protein] = quantile_upper
    df_ci.loc['theta', protein] = theta

print(df_ci)
```

**Выводы:** Нормальный доверительный интервал шире, чем центральный и квантильный интервалы, потому что нормальный интервал учитывает дисперсию бустрепных оценок.

## Задача 5

Бутстреп можно использовать для того, чтобы построить доверительные интервалы статистик, для которых у нас нет (или мы не знаем) хороших теорем об их распределениях. Рассмотрим коэффициент корреляции Пирсона.

$$r_{pearson}(X, Y) = \frac{\sum_{i=1}^n(X_i - \overline{X})(Y_i - \overline{Y})}{\sqrt{\sum_{i=1}^n(Y_i - \overline{Y})^2}\sqrt{\sum_{i=1}^n(Y_i - \overline{Y})^2}} = \frac{cov(X, Y)}{S_X \cdot S_Y},$$
где $X$ и $Y$ - векторы значений двух признаков длины $n$, $S$ - выборочные среднеквадратичные отклонения,  $cov$ - выборочная ковариация, $\overline{X}$ и $\overline{Y}$ - выборочные средние $X$ и $Y$. В питоне для подсчета выборочной корреляции можно использовать `sps.pearsonr()` или `np.corrcoef()`.

Чем ближе $r_{pearson}$ к $\pm1$, тем выше степень монотонной зависимости между признаками. На графике это выглядит так:

```python
sample_size = 500

X = sps.norm.rvs(size=sample_size)
Y = sps.norm.rvs(size=sample_size)
Z = 2 * X + sps.norm.rvs(size=sample_size)

r_weak = np.corrcoef(X, Y)[0, 1]
r_strong = np.corrcoef(X, Z)[0, 1]
```

```python
fig, axs = plt.subplots(ncols=2, figsize=(21, 6))

ax = axs[0]
sns.regplot(x=X, y=Y,
            scatter_kws={"color": "black"},
            line_kws={"color": "red"},
            ci=None,
            ax=ax)

ax.set_ylabel("Y")
ax.set_xlabel("X")
ax.set_title(f"Слабая корреляция r = {r_weak:.2f}")

ax = axs[1]
sns.regplot(x=X, y=Z,
            scatter_kws={"color": "black"},
            line_kws={"color": "red"},
            ci=None,
            ax=ax)

ax.set_ylabel("Z")
ax.set_xlabel("X")
ax.set_title(f"Сильная корреляция r = {r_strong:.2f}");
```

Реализуйте процедуру бутстрепа для корреляций между уровнями экспрессий каждой из пар белков: постройте гистограмму распределения бутстрепных коэффициентов корреляции и три вида доверительных интервалов. Результат оформите в виде графиков, сведенных в `plt.subplots()`, и общей таблицы со всеми доверительными интервалами (см. задание выше). Сделайте выводы.

```python
def correlation_estimator(sample):
    return np.corrcoef(sample[:, 0], sample[:, 1])[0, 1]
```

```python
df_ci = pd.DataFrame(columns=['Pair', 'normal_lower', 'normal_upper', 'central_lower', 'central_upper', 'quantile_lower', 'quantile_upper', 'theta'])


protein_pairs = [('ITSN1_N', 'DYRK1A_N'), ('ITSN1_N', 'ELK_N'), ('DYRK1A_N', 'ELK_N')]

fig, axs = plt.subplots(ncols=3, figsize=(21, 6))

for i, (protein1, protein2) in enumerate(protein_pairs):
    sample = np.column_stack((data[protein1].values, data[protein2].values))
    theta = correlation_estimator(sample)
    bootstrap_estimations = bootstrap(sample, correlation_estimator)

    normal_lower, normal_upper = bootstrap_normal_confidence_interval(theta, bootstrap_estimations)
    central_lower, central_upper = bootstrap_central_confidence_interval(theta, bootstrap_estimations)
    quantile_lower, quantile_upper = bootstrap_quantile_confidence_interval(theta, bootstrap_estimations)

    new_row = pd.DataFrame({
        'Pair': [f'{protein1} vs {protein2}'],
        'normal_lower': [normal_lower],
        'normal_upper': [normal_upper],
        'central_lower': [central_lower],
        'central_upper': [central_upper],
        'quantile_lower': [quantile_lower],
        'quantile_upper': [quantile_upper],
        'theta': [theta]
    })

    df_ci = pd.concat([df_ci, new_row], ignore_index=True)

    ax = axs[i]
    sns.histplot(bootstrap_estimations, kde=True, ax=ax)
    ax.set_title(f'Бутстрепные коэффициенты корреляции для {protein1} vs {protein2}')
    ax.axvline(normal_lower, color='r', linestyle='--', label='Стандартный')
    ax.axvline(normal_upper, color='r', linestyle='--')
    ax.axvline(central_lower, color='g', linestyle='-.', label='Центральный')
    ax.axvline(central_upper, color='g', linestyle='-.')
    ax.axvline(quantile_lower, color='b', linestyle=':', label='Квантильный')
    ax.axvline(quantile_upper, color='b', linestyle=':')
    ax.legend()

plt.tight_layout()
plt.show()

print(df_ci)
```

**Выводы:** по интервалам можем оценить уровень корреляции:
