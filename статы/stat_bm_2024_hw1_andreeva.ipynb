### Задача 1  
Создайте матрицу 10x10 по правилу:  
- если сумма индексов элемента четна - присваиваем элементу значение 0
- если сумма индексов элемента нечетна - присваиваем элементу значение 1

Используйте numpy. Визуализируйте полученную матрицу с помощью `plt.imshow()`.

```python
matrix = np.zeros((10, 10))

for i in range(10):
    for j in range(10):
        if (i + j) % 2 == 0:
            matrix[i, j] = 0
        else:
            matrix[i, j] = 1


plt.imshow(matrix)
plt.show()
```

### Задача 2
Изобразите график параметрически заданной функции. Переменные необходимо задавать с помощью Numpy.

$$ x=\alpha\times\left(\cos t+\frac{cos(\beta t)}{\beta}\right)$$

$$y=\alpha\times\left(\sin t−\frac{sin(\beta t)}{\beta}\right)$$

$$t∈[0;20π]$$

1. Реализуйте функции $x(t, \alpha, \beta)$ и $y(t, \alpha, \beta)$

```python
def x_func (t, a, b):
  x = a*(np.cos(t) + np.cos(b * t) / b)
  return x
def y_func(t, a, b):
  y = a*(np.sin(t) - np.sin(b * t) / b)
  return y
```

2. Постройте график в осях $x$ и $y$ при $\alpha=4$ и $\beta=1.1$. Подпишите оси.

```python
t = np.linspace(0, 20*np.pi, 100)
a = 4
b = 1.1
plt.figure(figsize=(4,4))
plt.plot(x_func(t, a, b), y_func(t, a, b))
plt.xlabel(r'$x$', fontsize=10)
plt.ylabel(r'$y$', fontsize=10)
plt.show()
```

3. Постройте графики для различных значений $\alpha$ и $\beta$.  
- $\alpha$ : [2, 4, 6, 8, 10]  
- $\beta$ : [1.2, 1.8, 2, 5, 10]  

У вас должна получиться сетка 5х5 графиков. На каждом графике укажите параметры $\alpha$ и $\beta$.

```python
a = np.array([2, 4, 6, 8, 10])
b = np.array([1.2, 1.8, 2, 5, 10])
t = np.linspace(0, 20*np.pi, 100)
fig, axes = plt.subplots(nrows = 5, ncols =5, figsize = (30,30))
for i in range(5):
  for j in range(5):
    axes[i,j].plot(x_func(t, a[i], b[j]), y_func(t, a[i], b[j]))
    axes[i,j].set(title= "при а = " + str(a[i]) + " и b = " + str(b[j]))
    plt.xlabel(r'$x$', fontsize=10)
    plt.ylabel(r'$y$', fontsize=10)
plt.show()
```

### Задача 3
- Создайте матрицу NхN, каждый элемент которой равен сумме индексов этого элемента, где N число из [10, 100, 1000, 10000]. Используйте циклы. Визуализируйте с помощью `plt.imshow()`. Для каждого N замерьте время работы вашего кода (это можно сделать с помощью библиотеки `time`).


```python
import time

def matrix_func(n):
    start_time = time.time()
    matrix = np.zeros((n, n))
    for i in range(n):
        for j in range(n):
            matrix[i, j] = i + j
    end_time = time.time()
    itog = end_time - start_time

    return matrix, itog

N_arr= [10, 100, 1000, 10000]
time_arr = []

for N in N_arr:
    matrix, itog_time = matrix_func(N)
    print("Матрица размера ", N,': ', itog_time, "секунд")
    time_arr.append(itog_time)
    plt.figure(figsize=(8, 8))
    plt.imshow(matrix)
    plt.title(f'Матрица {N}x{N}')
    plt.show()
```


- Сделайте вывод о временной сложности вашего алгоритма. Постройте график зависимости времени от N.


```python
plt.plot (N_arr, time_arr)
plt.xlabel("N")
plt.ylabel("t, с")
plt.title("Зависимость t от N")
plt.show()
```

Алгоритм имеет сложность $O(N^2)$, так как сначала создается матрица $N*N$, потом проходится двойной цикл, соответственно, затрачивается много времени.

- Повторите прогон вашего с алгоритма для N = 1000000. Что вы наблюдаете?

```python
def matrix_func(n):
    start_time = time.time()
    matrix = np.zeros((n, n))
    for i in range(n):
        for j in range(n):
            matrix[i, j] = i + j
    end_time = time.time()
    itog = end_time - start_time

    return matrix, itog


time_arr = []

N=1000000
matrix, itog_time = matrix_func(N)
print("Матрица размера ", N,': ', itog_time, "секунд")
time_arr.append(itog_time)
plt.figure(figsize=(8, 8))
plt.imshow(matrix)
plt.title(f'Матрица {N}x{N}')
plt.show()
```

### Задача 5

Выполните все шаги, описанные ниже и визуализируйте полученный результат. Результатом является квадратная матрица.

```python

# Создать матрицу a1 размера 3 x 3 из нулей
a1 = np.zeros((3, 3))

# Приравнять единице значения матрицы a1 по координатам (1, 1)
a1[1, 1] = 1

# Скопировать матрицу a1 в матрицу a2
a2 = a1.copy()

# Конкатенировать матрицы a1 и a2 по оси 1 и записать результат в матрицу a3
a3 = np.concatenate((a1, a2), axis=1)

# Создать матрицу a4 размера 6 из нулей
a4 = np.zeros(6)

# Запонить всё единицами кроме нулевой и последней ячейки в матрице a4
a4[1:-1] = 1

# Создать матрицу a5 размера 6 из нулей
a5 = np.zeros(6)

# Заполнить ячейки с координатами 2 и 3 единицами в матрице a5
a5[2:4] = 1

# Объединить матрицы a4 и a5 по нулевой оси и записать результат в матрицу a6
a6 = np.concatenate((a4, a5), axis=0)

# Создать матрицу a7 размера 6 из нулей
a7 = np.zeros(6)

# Конкатенировать матрицы a6 и a7 по нулевой оси и записать результат в матрицу a8
a8 = np.concatenate((a6, a7), axis=0)
a8 = a8.reshape((3, 6))

# Конкатенировать матрицы a3 и a8 по нулевой оси и записать результат в матрицу a9
a9 = np.concatenate((a3, a8), axis=0)

# Визуализировать матрицу a9 с помощью plt.imshow()
plt.imshow(a9,cmap='viridis')
plt.show()
```
