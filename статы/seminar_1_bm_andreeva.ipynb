## Задача 1 (<a  target="_blank" href="https://numpy.org/doc/stable/">numpy</a>):

1.
* Создайте массив чисел от $-2\pi$  до $2\pi$.

* Посчитайте сумму поэлементных квадратов синуса и косинуса для данного массива.

* С помощью `np.all` проверьте, что в ответе только единицы.

```python
arr = np.arange(-2*np.pi, 2*np.pi)
arr_new = []

for a in arr:
  arr_new.append(np.sin(a)**2 + np.cos(a)**2)

arr1 = np.array(arr_new)
result = np.all(arr1 == 1)

print(result)
```

2. Cоздайте матрицы $\begin{pmatrix} -3 & 4 \\ 4 & 3 \end{pmatrix}$ и $\begin{pmatrix} 2 & 1 \\ 1 & 2 \end{pmatrix}$. Посчитайте их поэлементное и матричное произведения.

```python
a = np.array([[-3, 4],[4,3]])
b = np.array([[2, 1],[1,2]])

print ("Матричное умножение: \n", a @ b)
print("Поэлементное умножение: \n" ,a * b)
```

3. В статистике и машинном обучении часто приходится иметь с функцией $RSS$, которая вычисляется по формуле $\sum_{i=1}^{n} (y_i - a_i)^2$, где $y_i$ — координаты одномерного вектора $y$,  $a_i$ — координаты одномерного вектора $a$. Посчитайте $RSS$ для $y = (1, 2, 3, 4, 5), a = (3, 2, 1, 0, -1)$.

```python
y = np.array([1, 2, 3, 4, 5])
a = np.array([3, 2, 1, 0, -1])

rss = np.sum((y - a) ** 2)

print("RSS:", rss)
```

## Задача 3 (<a  target="_blank" href="http://docs.scipy.org/doc/scipy/reference/stats.html">scipy.stats</a>):

${U}(a, b)$ — равномерное распределение.
  
Параметры в `scipy.stats`:  
* `loc` = $a$,
* `scale` = $b-a$.

Свойства распределения:
* математическое ожидание: $\frac{a+b}{2}$,
* дисперсия: $\frac{(b-a)^2}{12}$.

1. Постройте график плотности равномерного распределения ${U}(0, 1)$:

```python
x = np.linspace(-0.5, 1.5, 1000)
values = sps.uniform.pdf(x)
plt.plot(x, values, label='Плотность $\\mathcal{N}(0,1)$', lw=2)
plt.title("Плотность $U(0,1)$")
```

2. Сгенерируйте данные из равномерного распределения ${U}(0, 1)$ и сравните гистограмму с плотностью :

```python
x = np.linspace(-0.5, 1.5, 1000)
sample = sps.uniform.rvs(size=1000)  # выборка размера 1000

plt.figure(figsize=(16, 7))
plt.hist(sample, density=True, label='Гистограмма')
plt.plot(x, sps.uniform.pdf(x), lw=5, label='Плотность')
plt.legend()

plt.show()
```

3. Постройте плотности распределений ${U}(0, 1)$, ${U}(0, 2)$, ${U}(4, 1)$ на одном графике, опишите как плотность распределения зависит от его параметров:

```python
x = np.linspace(-2, 7, 1000)

plt.figure(figsize=(12, 6))
plt.plot(x, sps.uniform(0, 1).pdf(x), lw=5, label='Плотность U(0,1)')
plt.plot(x, sps.uniform(0, 2).pdf(x), lw=5, label='Плотность U(0,2)')
plt.plot(x, sps.uniform(4, 1).pdf(x), lw=5,  label='Плотность U(4,1)')

plt.legend()
plt.title('Плотности равномерного распределения')
plt.show()
```

**Ответ:** Можно заметить, что loc показывает начало интервала распределения. Scale указывает длительность этого интервала.

## Задача 4 (<a  target="_blank" href="https://pandas.pydata.org/docs/">pandas</a>):

Сгенерируйте случайную целочисленную матрицу $n \times m$, где $n=20, m=10$. Создайте из неё датафрейм, пронумеровав столбцы случайной перестановкой чисел из $\{1, \ldots, m\}$. Выберите столбцы с чётными номерами.

```python

n = 20
m = 10
matrix = np.random.randint(0, 100, size=(n, m))

df = pd.DataFrame(matrix)

random_columns = np.random.permutation(range(1, m + 1))
df.columns = random_columns

new_matrix = df.loc[:, df.columns % 2 == 0]

print("Исходная таблица:")
print(df)
print("\nТаблица с четными номерами столбцов:")
print(new_matrix)
```
